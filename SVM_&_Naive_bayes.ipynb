# SVM & Naive bayes

1. What is a Support Vector Machine (SVM)?
- A Support Vector Machine (SVM) is a supervised learning algorithm used for classification and regression. It finds the best boundary (called a hyperplane) that separates data into classes. SVM focuses on the points closest to the boundary, called support vectors, and tries to maximize the margin between classes. It can handle both linear and non-linear data using kernel functions.


2. What is the difference between Hard Margin and Soft Margin SVM?
- **Hard Margin SVM** requires that all data points be classified correctly with no errors. It's used when the data is perfectly separable.

- **Soft Margin SVM** allows some misclassifications to improve generalization. It's used when the data is not perfectly separable or has noise.

### In short:

* **Hard Margin**: No errors allowed, only for clean data.
* **Soft Margin**: Some errors allowed, better for real-world data.


3. What is the mathematical intuition behind SVM?
- The mathematical intuition behind SVM is to **maximize the margin** between two classes while **minimizing classification error**.

### Key Concepts:

1. **Hyperplane equation**:
   $w \cdot x + b = 0$,
   where $w$ is the weight vector and $b$ is the bias.

2. **Goal**:
   Maximize margin $= \frac{2}{\|w\|}$
   while ensuring all data points satisfy:

   * For class +1: $w \cdot x_i + b \geq 1$
   * For class -1: $w \cdot x_i + b \leq -1$

3. **Optimization problem**:

   $$
   \min \frac{1}{2} \|w\|^2 \quad \text{subject to } y_i(w \cdot x_i + b) \geq 1
   $$

4. **Soft margin adds slack variables** to allow errors:

   $$
   \min \frac{1}{2} \|w\|^2 + C \sum \xi_i \quad \text{subject to } y_i(w \cdot x_i + b) \geq 1 - \xi_i
   $$

Where:

* $C$: penalty parameter for misclassification.
* $\xi_i$: slack variable for errors.

In short: **SVM tries to find a boundary with the largest gap between classes while minimizing misclassification.**


4. What is the role of Lagrange Multipliers in SVM?
- In SVM, **Lagrange multipliers** are used to solve the **constrained optimization problem** by converting it into a **dual problem**.

### Role of Lagrange Multipliers:

1. They help include the constraints $y_i(w \cdot x_i + b) \geq 1$ into the optimization.
2. The SVM problem is reformulated using a **Lagrangian** function:

   $$
   L(w, b, \alpha) = \frac{1}{2} \|w\|^2 - \sum_{i=1}^{n} \alpha_i [y_i(w \cdot x_i + b) - 1]
   $$
3. **$\alpha_i$** are the Lagrange multipliers—each corresponds to a training example.
4. Only points with **non-zero $\alpha_i$** (support vectors) affect the final decision boundary.
5. Solving the dual form is computationally more efficient, especially with kernels.

**In short**: Lagrange multipliers help solve the SVM optimization by focusing on the support vectors and simplifying the computation.


5. What are Support Vectors in SVM?
- **Support Vectors** are the data points in SVM that lie closest to the decision boundary (hyperplane).

They are the **most critical points** because:

* They **define the position and orientation** of the hyperplane.
* The margin is calculated based on them.
* Only these points have **non-zero Lagrange multipliers**.

**In short**:
Support vectors are the key data points that directly influence the SVM model’s decision boundary.


6. What is a Support Vector Classifier (SVC)?
- A **Support Vector Classifier (SVC)** is the practical implementation of SVM for classification tasks.

- It finds the best decision boundary (hyperplane) that separates different classes by maximizing the margin and using support vectors.

**In short**:
SVC is an SVM-based model used to classify data by finding the optimal separating boundary between classes.


7. What is a Support Vector Regressor (SVR)?
- A **Support Vector Regressor (SVR)** is the regression version of SVM.

- It tries to find a function that fits the data **within a certain margin (ε)** and allows some errors beyond this margin using slack variables.

**In short**:
SVR predicts continuous values by finding a line (or curve) that best fits the data while keeping errors within a tolerable range.


8. What is the Kernel Trick in SVM?
- The **Kernel Trick** in SVM allows it to handle **non-linear data** by **transforming** it into a higher-dimensional space where it becomes linearly separable—**without actually computing the transformation**.

### Common Kernels:

* **Linear**: $K(x, x') = x \cdot x'$
* **Polynomial**: $K(x, x') = (x \cdot x' + c)^d$
* **RBF (Gaussian)**: $K(x, x') = \exp(-\gamma \|x - x'\|^2)$

**In short**:
The kernel trick lets SVM solve non-linear problems by using kernel functions to compute dot products in high-dimensional space efficiently.


9. Compare Linear Kernel, Polynomial Kernel, and RBF Kernel
- Here's a short comparison of **Linear**, **Polynomial**, and **RBF** kernels in SVM:

| **Kernel**         | **Equation**                            | **Use Case**                     | **Pros**                       | **Cons**                           |
| ------------------ | --------------------------------------- | -------------------------------- | ------------------------------ | ---------------------------------- |
| **Linear**         | $K(x, x') = x \cdot x'$                 | Linearly separable data          | Fast, simple, fewer parameters | Poor with non-linear data          |
| **Polynomial**     | $K(x, x') = (x \cdot x' + c)^d$         | Moderately complex relationships | Handles some non-linearity     | Slower, needs tuning of degree $d$ |
| **RBF (Gaussian)** | $K(x, x') = \exp(-\gamma \|x - x'\|^2)$ | Complex, non-linear data         | Highly flexible, powerful      | Prone to overfitting if not tuned  |

**In short**:

* **Linear**: Best for simple, linearly separable data.
* **Polynomial**: Adds flexibility for moderate non-linearity.
* **RBF**: Most powerful for complex patterns, but needs careful tuning.


10.  What is the effect of the C parameter in SVM?
- The **C parameter** in SVM controls the **trade-off between margin size and classification error**.

* **High C**:

  * The model tries to classify all training points correctly (low tolerance for errors).
  * Results in a **narrow margin** and can lead to **overfitting**.

* **Low C**:

  * The model allows some misclassifications to get a **wider margin**.
  * Leads to better **generalization** but possibly more training errors.

**In short:**
**C controls how much you want to avoid misclassifying training data vs. having a simpler, wider margin.**


11. What is the role of the Gamma parameter in RBF Kernel SVM?
- The **Gamma** parameter in an RBF kernel controls the **influence range** of a single training example.

* **High Gamma**:

  * Each point’s influence is **very close** (narrow).
  * The decision boundary becomes **more complex**, fitting tightly to data (risk of overfitting).

* **Low Gamma**:

  * Each point’s influence is **farther** (wide).
  * The decision boundary is **smoother** and less sensitive to individual points (risk of underfitting).

**In short:**
Gamma controls how much each training point affects the shape of the decision boundary in RBF kernel SVM.


12. What is the Naïve Bayes classifier, and why is it called "Naïve"?
- The **Naïve Bayes classifier** is a simple probabilistic machine learning model used for classification. It applies **Bayes’ theorem** assuming that all features are **independent** of each other given the class.

- It's called **"Naïve"** because it **naively assumes feature independence**, which is often not true in real data, but despite this simplification, it works well in many cases.

**In short:**
Naïve Bayes classifies data based on probabilities and assumes features don’t influence each other, which is why it’s called "naïve."


13. What is Bayes’ Theorem?
- **Bayes’ Theorem** is a formula that describes how to update the probability of a hypothesis based on new evidence.

Mathematically:

$$
P(A|B) = \frac{P(B|A) \times P(A)}{P(B)}
$$

Where:

* $P(A|B)$ = Probability of event A given B (posterior)
* $P(B|A)$ = Probability of event B given A (likelihood)
* $P(A)$ = Probability of event A (prior)
* $P(B)$ = Probability of event B (evidence)

**In short:**
Bayes’ Theorem helps update beliefs about A after seeing B.


14. Explain the differences between Gaussian Naïve Bayes, Multinomial Naïve Bayes, and Bernoulli Naïve Bayes.
- Here’s a quick comparison of the three Naïve Bayes types:

| **Type**                    | **Data Type**           | **Use Case**                              | **How It Works**                                                               |
| --------------------------- | ----------------------- | ----------------------------------------- | ------------------------------------------------------------------------------ |
| **Gaussian Naïve Bayes**    | Continuous numeric data | Features like height, weight, temperature | Assumes features follow a **normal (Gaussian)** distribution.                  |
| **Multinomial Naïve Bayes** | Discrete count data     | Text classification (word counts)         | Uses feature counts, models **frequency of events**.                           |
| **Bernoulli Naïve Bayes**   | Binary/Boolean data     | Text classification with presence/absence | Works on **binary features** indicating presence (1) or absence (0) of events. |

**In short:**

* Gaussian for continuous data.
* Multinomial for count data (e.g., word counts).
* Bernoulli for binary data (e.g., word present or not).


15. When should you use Gaussian Naïve Bayes over other variants?
- Use **Gaussian Naïve Bayes** when your features are **continuous numerical values** that roughly follow a **normal (Gaussian) distribution**.

### Examples:

* Predicting if a patient has a disease based on blood pressure, age, or temperature.
* Any problem with real-valued measurements.

**In short:**
Choose Gaussian Naïve Bayes when your data is continuous and not categorical or binary.


16.  What are the key assumptions made by Naïve Bayes?
- Key assumptions of Naïve Bayes are:

1. **Feature Independence:** All features are assumed to be **independent** of each other given the class label.
2. **Equal Importance:** Each feature contributes **equally and independently** to the outcome.
3. **Feature Distribution:** The features follow a certain distribution depending on the variant (e.g., Gaussian for continuous data).

**In short:**
Naïve Bayes assumes features don’t influence each other and follow specific distributions.


17. What are the advantages and disadvantages of Naïve Bayes?
**Advantages of Naïve Bayes:**

* Simple and fast to train and predict.
* Works well with high-dimensional data.
* Performs well even with small training data.
* Effective for text classification and spam filtering.
* Handles both continuous and discrete data (with variants).

**Disadvantages of Naïve Bayes:**

* Assumes feature independence, which is often unrealistic.
* Can perform poorly if features are highly correlated.
* Zero probability problem if a feature-category pair never appears in training (can be fixed with smoothing).
* Less flexible compared to more complex models.

**In short:**
Naïve Bayes is fast and easy but relies on a strong independence assumption that may limit accuracy.


18. Why is Naïve Bayes a good choice for text classification?
- Naïve Bayes is great for text classification because:

* It handles **high-dimensional data** (like thousands of words) efficiently.
* Works well with **sparse data**, where many words don’t appear in every document.
* The **independence assumption** roughly fits, since word presence can be treated as mostly independent.
* It’s **fast to train and predict**, making it scalable for large text datasets.
* Performs surprisingly well even with simple assumptions.

**In short:**
Naïve Bayes is fast, scalable, and effective for classifying text despite its simplicity.


19. Compare SVM and Naïve Bayes for classification tasks.
- Here’s a quick comparison between **SVM** and **Naïve Bayes** for classification:

| Aspect                   | SVM                                                   | Naïve Bayes                                                   |
| ------------------------ | ----------------------------------------------------- | ------------------------------------------------------------- |
| **Approach**             | Finds optimal boundary (margin) between classes       | Probabilistic, based on Bayes’ theorem                        |
| **Assumptions**          | No strong feature independence assumption             | Assumes feature independence                                  |
| **Performance**          | Often better with complex, high-dimensional data      | Works well with simple, high-dimensional data                 |
| **Data Type**            | Handles both linear and non-linear data using kernels | Handles categorical, continuous, or binary data with variants |
| **Training Speed**       | Slower, especially with large datasets                | Very fast, even with large datasets                           |
| **Interpretability**     | Harder to interpret (especially with kernels)         | Easier to interpret (probabilities)                           |
| **Sensitivity to Noise** | Less sensitive with soft margin tuning                | More sensitive if independence assumption is violated         |
| **Use Cases**            | Image recognition, bioinformatics, complex patterns   | Text classification, spam filtering, simple tasks             |

**In short:**

* Use **SVM** for complex, non-linear data and when accuracy is critical.
* Use **Naïve Bayes** for fast, simple, and high-dimensional tasks like text classification.


20.  How does Laplace Smoothing help in Naïve Bayes?
- **Laplace Smoothing** helps Naïve Bayes by **avoiding zero probabilities** for features that don’t appear in the training data for a class.

- Without smoothing, if a feature never occurs with a class, its probability becomes zero, which can make the whole prediction zero (due to multiplication).

- Laplace smoothing adds a small constant (usually 1) to all feature counts, ensuring no probability is zero.

**In short:**
It prevents zero probabilities and improves model robustness, especially with rare features.


```python
# 21. Write a Python program to train an SVM Classifier on the Iris dataset and evaluate accuracy.
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split dataset into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Create SVM classifier
svm_classifier = SVC(kernel='linear')  # Using linear kernel

# Train the classifier
svm_classifier.fit(X_train, y_train)

# Predict on test set
y_pred = svm_classifier.predict(X_test)

# Evaluate accuracy
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy of SVM classifier on Iris dataset: {accuracy:.2f}")
```

```python
#24. Write a Python program to train an SVM Classifier with a Polynomial Kernel and visualize the decision boundary.
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Load Iris dataset and select first two features for 2D visualization
iris = datasets.load_iris()
X = iris.data[:, :2]  # only first two features
y = iris.target

# Use only two classes for binary classification (e.g., class 0 and 1)
X = X[y != 2]
y = y[y != 2]

# Create and train SVM classifier with polynomial kernel
svm_poly = SVC(kernel='poly', degree=3, coef0=1, C=1)
svm_poly.fit(X, y)

# Create a grid to plot decision boundary
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

# Predict on the grid points
Z = svm_poly.predict(np.c_[xx.ravel(), yy.ravel()])
Z = Z.reshape(xx.shape)

# Plot decision boundary and data points
plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
plt.xlabel('Feature 1')
plt.ylabel('Feature 2')
plt.title('SVM with Polynomial Kernel (degree=3)')
plt.show()
```

```python
#27. Write a Python program to train an SVM Classifier with different C values and compare the decision boundaries visually.
import numpy as np
import matplotlib.pyplot as plt
from sklearn import datasets
from sklearn.svm import SVC

# Load Iris dataset and select first two features for 2D visualization
iris = datasets.load_iris()
X = iris.data[:, :2]
y = iris.target

# Use only two classes for binary classification
X = X[y != 2]
y = y[y != 2]

# Different C values to compare
C_values = [0.1, 1, 10, 100]

# Create a grid to plot decision boundaries
x_min, x_max = X[:, 0].min() - 1, X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1, X[:, 1].max() + 1
xx, yy = np.meshgrid(np.linspace(x_min, x_max, 500),
                     np.linspace(y_min, y_max, 500))

plt.figure(figsize=(16, 12))

for i, C in enumerate(C_values, 1):
    svm = SVC(kernel='linear', C=C)
    svm.fit(X, y)

    Z = svm.predict(np.c_[xx.ravel(), yy.ravel()])
    Z = Z.reshape(xx.shape)

    plt.subplot(2, 2, i)
    plt.contourf(xx, yy, Z, alpha=0.3, cmap=plt.cm.coolwarm)
    plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.coolwarm, edgecolors='k')
    plt.title(f"SVM with Linear Kernel (C={C})")
    plt.xlabel('Feature 1')
    plt.ylabel('Feature 2')

plt.tight_layout()
plt.show()
```

```python
#29. Write a Python program to apply feature scaling before training an SVM model and compare results with unscaled data.
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.preprocessing import StandardScaler
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# --- Without scaling ---
svm_unscaled = SVC(kernel='rbf')
svm_unscaled.fit(X_train, y_train)
y_pred_unscaled = svm_unscaled.predict(X_test)
acc_unscaled = accuracy_score(y_test, y_pred_unscaled)

# --- With feature scaling ---
scaler = StandardScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)

svm_scaled = SVC(kernel='rbf')
svm_scaled.fit(X_train_scaled, y_train)
y_pred_scaled = svm_scaled.predict(X_test_scaled)
acc_scaled = accuracy_score(y_test, y_pred_scaled)

print(f"Accuracy without scaling: {acc_unscaled:.2f}")
print(f"Accuracy with scaling: {acc_scaled:.2f}")
```

```python
#40. Write a Python program to perform Recursive Feature Elimination (RFE) before training an SVM Classifier and compare accuracy.
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.feature_selection import RFE
from sklearn.metrics import accuracy_score

# Load Iris dataset
iris = datasets.load_iris()
X = iris.data
y = iris.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train SVM without feature selection
svm = SVC(kernel='linear', random_state=42)
svm.fit(X_train, y_train)
y_pred = svm.predict(X_test)
acc_before = accuracy_score(y_test, y_pred)

# Apply Recursive Feature Elimination (select 2 features)
rfe = RFE(estimator=SVC(kernel='linear'), n_features_to_select=2)
X_train_rfe = rfe.fit_transform(X_train, y_train)
X_test_rfe = rfe.transform(X_test)

# Train SVM with selected features
svm_rfe = SVC(kernel='linear', random_state=42)
svm_rfe.fit(X_train_rfe, y_train)
y_pred_rfe = svm_rfe.predict(X_test_rfe)
acc_after = accuracy_score(y_test, y_pred_rfe)

print(f"Accuracy before RFE: {acc_before:.2f}")
print(f"Accuracy after RFE (2 features): {acc_after:.2f}")
```

```python
#45. Write a Python program to train a Naïve Bayes classifier and evaluate its performance using the ROC-AUC score.
from sklearn import datasets
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import GaussianNB
from sklearn.metrics import roc_auc_score, roc_curve
import matplotlib.pyplot as plt

# Load Breast Cancer dataset (binary classification)
data = datasets.load_breast_cancer()
X = data.data
y = data.target

# Split data
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train Gaussian Naive Bayes
gnb = GaussianNB()
gnb.fit(X_train, y_train)

# Predict probabilities for the positive class
y_prob = gnb.predict_proba(X_test)[:, 1]

# Calculate ROC-AUC score
auc = roc_auc_score(y_test, y_prob)
print(f"ROC-AUC Score: {auc:.2f}")

# Optional: Plot ROC curve
fpr, tpr, thresholds = roc_curve(y_test, y_prob)
plt.plot(fpr, tpr, label=f'ROC curve (area = {auc:.2f})')
plt.plot([0, 1], [0, 1], 'k--')  # diagonal line
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC Curve - Gaussian Naive Bayes')
plt.legend(loc='lower right')
plt.show()
```
